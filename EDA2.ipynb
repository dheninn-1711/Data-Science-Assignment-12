```python
#import library
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Select numerical features
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
numerical_features
```

```python
# Standard Scaling
standard_scaler = StandardScaler()
stand_data=data[numerical_features] = (standard_scaler.fit_transform(data[numerical_features]))
stand_data
```

```python
# Min-Max Scaling
minmax_scaler = MinMaxScaler()
min_max_data=data[numerical_features] = (minmax_scaler.fit_transform(data[numerical_features]))
min_max_data
```

Standard Scaling: Centers the data around zero with a unit variance. Preferred when the data follows a Gaussian distribution.

Min-Max Scaling: Scales data to a range of [0,1]. Useful when features have different ranges and no Gaussian distribution.

```python
#One-Hot Encoding for categorical variables with less than 5 categories
data = pd.get_dummies(data, columns=['marital_status','relationship','race','sex'], drop_first=True)
data
```

```python
#import library
from sklearn.preprocessing import LabelEncoder
```

```python
#label-encoding for categorical variables with more than 5 categories
label_encoder = LabelEncoder()
data['workclass'] = label_encoder.fit_transform(data['workclass'])
data['education'] = label_encoder.fit_transform(data['education'])
data['occupation'] = label_encoder.fit_transform(data['occupation'])
data['native_country'] = label_encoder.fit_transform(data['native_country'])
data
```

Pros and Cons:

One-Hot Encoding:

Pros: No ordinal relationship assumed, suitable for algorithms that cannot handle categorical data.

Cons: Increases dimensionality, can lead to sparse matrices.

Label Encoding:

Pros: Simple, does not increase dimensionality.

Cons: Assumes ordinal relationship, which may not be appropriate for all categorical data.

```python
#import libraries
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings('ignore')
```

```python
#identify and remove outliers
iso_forest = IsolationForest(contamination=0.05)
outliers = iso_forest.fit_predict(data[numerical_features])
data = data[outliers == 1]
data
```

Outliers can skew statistical measures (mean and variance), leading to inaccurate model assumptions and potentially causing overfitting, which results in poor generalization to new data.

In distance-based algorithms, outliers can distort distance calculations, leading to incorrect classifications or cluster assignments, ultimately compromising the model's accuracy and reliability.
